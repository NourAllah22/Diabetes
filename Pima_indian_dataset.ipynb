```python
import pandas as pd #data frame library
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

### Load and review data

```python
df=pd.read_csv('diabetes.csv') #load Pima data
df.head() #to show the beginning row of our dataset
```

df.info()

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Find missing values
print('Missing values:{}'.format(df.isnull().any().sum()))
# Find duplicated records
print('\nNumber of duplicated records: {}'.format(df.duplicated().sum()))
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
```

```python
# Normalization  to update data in range(0,1)
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(X)
print(scaled_data)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_data,Y,test_size=0.2,random_state=0)
```

###Model




```python
model=LogisticRegression()
# Train model
model.fit(x_train,y_train)
```

```python
print(model.score(x_test,y_test))
```

```python
data = model.predict([[6,148,72,35,0,50,33.6,0.627]])
for i in range (1):
  if(data[i]==0):
    print("Non-diabetic")
  else:
    print("Diabetic")
print(data)
```
